```python
# =========================================
# SIMULASI TEKANAN RAM (VIRTUAL MEMORY)
# Platform: Google Colab
# =========================================

import psutil
import time
import matplotlib.pyplot as plt

data = []
ram_usage = []
step = 0

print("=== INFORMASI RAM AWAL ===")
ram = psutil.virtual_memory()
print(f"Total RAM     : {ram.total / (1024**3):.2f} GB")
print(f"Available RAM : {ram.available / (1024**3):.2f} GB")
print(f"Usage         : {ram.percent}%")
print("=========================\n")

try:
    while True:
        # Alokasi ~50 MB per iterasi
        data.append('A' * 50_000_000)
        step += 1

        ram = psutil.virtual_memory()
        ram_usage.append(ram.percent)

        print(f"Step {step} | RAM Used: {ram.percent}% | Available: {ram.available / (1024**3):.2f} GB")

        # Hentikan manual kalau sudah tinggi
        if ram.percent > 90:
            print("\nRAM hampir penuh â†’ simulasi working set > RAM")
            break

        time.sleep(0.5)

except MemoryError:
    print("\nMemoryError: RAM tidak mencukupi (simulasi thrashing ekstrem)")

# ===============================
# VISUALISASI HASIL
# ===============================
plt.figure()
plt.plot(ram_usage)
plt.xlabel("Waktu (step)")
plt.ylabel("RAM Usage (%)")
plt.title("Simulasi Tekanan RAM (Virtual Memory Concept)")
plt.show()
```
